```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
y_2=sns.load_dataset("titanic").head(100)
fig,axs=plt.subplots(nrows=2,ncols=2, figsize=(10,8))
sns.lineplot(x="survived",y="pclass",data=y_2,hue="sex",size=5,style="sex",palette='turbo',markers=["o","*"],dashes=True,legend=True,ax=axs[0,0])
plt.grid()
axs[0,0].set_title("TITANIC1")
sns.lineplot(x="age",y="sibsp",data=y_2,hue="parch",size=5,style="parch",palette='Accent',markers=["o","*","<",">","^"],dashes=True,legend=True,ax=axs[0,1])
plt.grid()
axs[0,1].set_title("TITANIC2")
sns.lineplot(x="fare",y="embarked",data=y_2,hue="class",size=5,style="class",palette='turbo',markers=["o","*","<"],dashes=True,legend=True,ax=axs[1,0])
plt.grid()
axs[1,0].set_title("TITANIC3")
sns.lineplot(x="adult_male",y="deck",data=y_2,hue="embark_town",size=5,style="embark_town",palette='Accent',markers=["o","*","<"],dashes=True,legend=True,ax=axs[1,1])
plt.grid()
axs[1,1].set_title("TITANIC4")
fig.suptitle('Multiple plots with Seaborn')
plt.subplots_adjust(hspace=0.3)
plt.savefig("myplot.png")
```
