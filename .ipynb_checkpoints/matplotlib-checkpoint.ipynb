```python
import matplotlib.pyplot as plt
import pandas as pd
data=pd.read_excel('data_graph.xlsx')
fig,axs=plt.subplots(nrows=2,ncols=2, figsize=(10,8))
x_axis1 = data['FRUITS']
y_axis1 = data['PRICE']
x_axis2 = data['VEGETABLES']
y_axis2 = data['PRICE']
x_axis3 = data['FRUITS']
y_axis3 = data['QUANTITY']
x_axis4 = data['VEGETABLES']
y_axis4 = data['QUANTITY']
plt.subplot(2, 2, 1)
plt.plot(x_axis1, y_axis1)
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.subplot(2, 2, 2)
plt.plot(x_axis2, y_axis2)
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.subplot(2, 2, 3)
plt.plot(x_axis3, y_axis3)
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.subplot(2, 2, 4)
plt.plot(x_axis4, y_axis4)
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.show()
#plt.savefig("myplot2.png")
```
