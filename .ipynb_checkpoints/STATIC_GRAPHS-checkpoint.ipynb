```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
y_2=sns.load_dataset("titanic").head(100)
sns.lineplot(x="survived",y="pclass",data=y_2,hue="sex",size=5,style="sex",palette='turbo',markers=["o","*"],dashes=True,legend=True)
plt.grid()
plt.title("TITANIC1")
plt.show()
sns.lineplot(x="age",y="sibsp",data=y_2,hue="parch",size=5,style="parch",palette='Accent',markers=["o","*","<",">","^"],dashes=True,legend=True)
plt.grid()
plt.title("TITANIC2")
plt.show()
sns.lineplot(x="fare",y="embarked",data=y_2,hue="class",size=5,style="class",palette='turbo',markers=["o","*","<"],dashes=True,legend=True)
plt.grid()
plt.title("TITANIC3")
plt.show()
sns.lineplot(x="adult_male",y="deck",data=y_2,hue="embark_town",size=5,style="embark_town",palette='Accent',markers=["o","*","<"],dashes=True,legend=True)
plt.grid()
plt.title("TITANIC4")
plt.show()
sns.set(rc={"figure.figsize": (6,6)})
plt.savefig("myplot.png")
```
